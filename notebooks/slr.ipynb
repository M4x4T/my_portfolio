**What is Simple Linear Regression?**

Simple Regression is a statistical method to build a predictive model for a dependent variable "y." More simply put, it is a way to analyze and quantify the relationship between one independent variable and one dependent variable.

Let me make this concrete. Imagine we have a dataset with just 2 columns: the **price of a house** and the **size of a house**. Our goal is to predict the price based on the size. Mathematically, we express this as:

**f(x) = y = mx + b**

But here comes the real question — what exactly are **"m"** and **"b"**?

**The Math Behind the Line**

In mathematical terms, linear regression is an **approximation line** drawn between two variables. The equation of that line is **y = mx + b**, where:
- **m** is the slope — how much y changes for every unit increase in x
- **b** is the intercept — where the line crosses the y-axis

The goal of the model is to find the values of m and b that produce the most accurate predictions. To do that, we need to measure how wrong our model is — and that is where the **loss function** comes in.

**The Loss Function and OLS**

The loss function **J(m, b)** measures the total error of our model. Mathematically it looks like this:

**J(m, b) = Σ(y − ŷ)²**

Where **y** is the actual value and **ŷ** is the predicted value. We square the differences to penalize larger errors more heavily and to avoid negative and positive errors cancelling each other out.

Now — how do we minimize this? We use a method called **Ordinary Least Squares (OLS)**, which finds the exact values of m and b that minimize the sum of squared errors. No guessing, just math.


### Here is the example slr implementation in python

```python
# instalations
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Defining an example dataset and fit a linear regression model

```python

# Dataset: size of house (m²) and price ($1000s)
size = np.array([50, 70, 90, 110, 130, 150, 170]).reshape(-1, 1)
price = np.array([150, 200, 250, 290, 340, 400, 450])

# Build and train the model
model = LinearRegression()
model.fit(size, price)

m = model.coef_[0]
b = model.intercept_

print(f"Slope     (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Equation     : y = {m:.2f}x + {b:.2f}")
```

### Checking metric and visualizing

```python

print(f"R² Score     : {model.score(size, price):.2f}")

# Predict price for a 120 m² house
predicted = model.predict([[120]])
print(f"\nPredicted price for 120 m² house: ${predicted[0]:.1f}k")

# Visualize
plt.scatter(size, price, color='blue', label='Actual data')
plt.plot(size, model.predict(size), color='red', label='Regression line')
plt.xlabel('Size (m²)')
plt.ylabel('Price ($1000s)')
plt.title('Simple Linear Regression — House Price Prediction')
plt.legend()
plt.show()
```


**One Important Limitation**

Simple Linear Regression assumes a **linear relationship** between your variables. If the real relationship is curved or more complex, this model will perform poorly no matter how well you tune it. Always visualize your data before applying the model — a simple scatter plot can save you a lot of time.

**Final Thought**

Simple Linear Regression is one of the most elegant tools in machine learning — not because it is powerful, but because it is interpretable. You can look at the equation and immediately understand what the model learned. In a field where models are getting increasingly complex and opaque, that clarity still has real value.
